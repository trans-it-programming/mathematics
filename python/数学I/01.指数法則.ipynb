# 指数法則

$$
    a^{n} = a \times a \times \cdots \times a
$$

```python
import sympy
```

```python
# 変数 a, b の定義
a = sympy.Symbol('a')
b = sympy.Symbol('b')
a, b
```

```python
# a を 2回掛ける
a * a
```

```python
# a を 6回掛ける
a * a * a * a * a * a
```

### 指数に関する公式
$m$, $n$: 自然数、$a\neq 0$

- $a^{0} = 1$
- $a^{1} = a$
- $a^{m} \times a^{n} = a^{m+n}$
- $(a^{m})^{n}=a^{mn}$
- ${\displaystyle \frac{a^{m}}{a^{n}}} = a^{m - n}$
- ${\displaystyle \left(\frac{b}{a}\right)^{m} = \frac{b^{m}}{a^{m}}}$
- $(a \times b)^{m} = a^{m} \times b^{m}$

##### 例

- $a^{0} = 1$

```python
# 0乗は1になる
a ** 0
```

- $a^{1} = a$

```python
# 1乗はその数自身
a ** 1
```

- $a^{m} \times a^{n} = a^{m + n}$

```python
# m = 5, n = 3
(a ** 5) * (a ** 3)
```

```python
# m = 19, n = 10
(a ** 19) * (a ** 10)
```

- $(a^{m})^{n}=a^{mn}$

```python
# m = 5, n = 3
(a ** 5) ** 3
```

```python
# m = 19, n = 10
(a ** 19) ** 10
```

- ${\displaystyle \frac{a^{m}}{a^{n}}} = a^{m - n}$

```python
# m = 5, n = 3
(a ** 5) / (a ** 3)
```

```python
# m = 19, n = 10
(a ** 19) / (a ** 10)
```

- ${\displaystyle \left(\frac{b}{a}\right)^{m} = \frac{b^{m}}{a^{m}}}$

```python
(b / a) ** 3
```

```python
(b / a) ** 10
```

- $(a \times b)^{m} = a^{m} \times b^{m}$

```python
(a * b) ** 4
```

```python
(a * b) ** 13
```
